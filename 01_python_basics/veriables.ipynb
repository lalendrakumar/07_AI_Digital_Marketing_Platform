```python
# 01. Variable assignment
a=10
```

```python
# 02. Assigning Different Types of Variables
b=3.14
c="Hello, World!"
d=[1, 2, 3, 4, 5]
e=(10, 20, 30)
f={'name': 'Alice', 'age': 30}
```

```python
# 03. Printing Variables
print('Age:', a)
print('Pi:', b)
print('Greeting:', c)
print('List:', d)
print('Tuple:', e)
print('Dictionary:', f)
```

```python
# 04. Modern Alternative (f-strings)
print(f'Age: {a}')
print(f'pi: {b}')
print(f'Greeting: {c}')
print(f'List: {d}')
print(f'Tuple: {e}')
print(f'Dictionary: {f}')
```

```python
# 04. Naming Conventions
my_variable = 42          # Snake case
myVariable = 42          # Camel case
MyVariable = 42          # Pascal case
MY_VARIABLE = 42         # Upper case with underscores
print(my_variable, myVariable, MyVariable, MY_VARIABLE)
```

```python
# 05. Case Sensitivity
Var = 100
var = 200
print("Var:", Var)
print("var:", var)
```

```python
# 06. Dynamic Typing
x = 10
print("x is:", x, "of type", type(x))
```

```python
# 07. Type Checking
print("Is x an integer?", isinstance(x, int))
x = "Now I'm a string"
print("x is:", x, "of type", type(x))
print("Is x a string?", isinstance(x, str))
```

```python
# 08. Type Conversion
num_str = "123"
num_int = int(num_str)
print("Converted string to integer:", num_int)
num_float = float(num_int)
print("Converted integer to float:", num_float)
bool_value = bool(num_int)
print("Converted integer to boolean:", bool_value)
```

```python
# Input Function
user_input = input("Enter something: ")
print("You entered:", user_input)
```

```python
# Correct Conversion
age_str = input("Enter your age: ")
age_int = int(age_str)
print("You are", age_int, "years old.")
```

```python
# Simple Calculator
num1_str = input("Enter first number: ")
num2_str = input("Enter second number: ")
num1 = float(num1_str)
num2 = float(num2_str)
sum_result = num1 + num2
print("The sum is:", sum_result)
```

```python
#Improved Safe Version
try:
    num1_str = input("Enter first number: ")
    num2_str = input("Enter second number: ")
    num1 = float(num1_str)
    num2 = float(num2_str)
    sum_result = num1 + num2
    print("The sum is:", sum_result)
except ValueError:
    print("Invalid input! Please enter numeric values.")
```
